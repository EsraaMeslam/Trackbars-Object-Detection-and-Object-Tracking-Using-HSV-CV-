```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Object Detection and Object Tracking Using HSV

```python
frame=cv2.imread('ball.jpg',-1) #read the image,-1 for alpha channel (no change in color)

#HSV >>HUE , SATURATION ,VALUE
hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV) #cvt the image to hsv

#detect upper,lower boundary
lower_b=np.array([110,120,20])
upper_b=np.array([150,255,255])
    
mask=cv2.inRange(hsv,lower_b,upper_b)#(Mask)threshold the HSV (make all image black except ths range)
    
res=cv2.bitwise_and(frame,frame,mask=mask)

plt.figure(figsize=(20, 10))

plt.subplot(1, 3, 1)
plt.imshow(frame,cmap='gray')
plt.title('detect Orange balls')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(mask,cmap='gray')
plt.title('mask')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(res,cmap='gray')
plt.title('Result')
plt.axis('off')

plt.show()
```
